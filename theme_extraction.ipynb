# Theme Extraction Using Text Mining Techniques

## Introduction
Using the CSV output of the web scraper (article_scraper.ipynb), this simple script aims to identify recurring themes or topics in the articles. We'll use text mining techniques such as topic modeling to achieve this. Specifically, we'll use Latent Dirichlet Allocation (LDA), a popular method for topic modeling.

## Explanation of Latent Dirichlet Allocation (LDA)
Latent Dirichlet Allocation (LDA) is a generative statistical model widely used in natural language processing (NLP) for topic modeling. It helps to discover the hidden topics present in a collection of documents. Here‚Äôs a detailed explanation of how LDA works:

## Conceptual Overview:

### Documents and Words:

- LDA assumes that documents are a mixture of topics, and topics are a mixture of words.
- Each document can be represented as a distribution over topics.
- Each topic can be represented as a distribution over words.

###  Generative Process:

- LDA is based on a generative process for creating documents:
1. Choose a topic distribution for each document from a Dirichlet distribution.
2. For each word in the document:
- Select a topic from the document‚Äôs topic distribution.
- Choose a word from the topic‚Äôs word distribution.

### Dirichlet Distribution:

- A Dirichlet distribution is used to generate distributions over a finite set (like topics or words).
- It is parameterized by a vector of positive reals which can influence the sparsity of the distribution.

## Steps in LDA:

### Initialization:

- Define the number of topics 
ùêæ

- Initialize the topic distributions for each document and the word distributions for each topic.

### E-Step (Expectation):

- For each document, estimate the probability of each topic for each word.
- This involves calculating the likelihood that a given word in a document is generated by each topic.

### M-Step (Maximization):

- Adjust the topic distributions for each document and the word distributions for each topic to maximize the likelihood of the observed data.
- This involves updating the topic-word and document-topic distributions based on the current word-topic assignments.

### Iterate:

- Repeat the E-Step and M-Step until convergence, where the topic distributions stabilize.

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

## 1. Load the Data:

Reads the CSV file into a pandas DataFrame.

```python
# Step 1: Load the CSV Data
file_path = 'scraped_articles.csv'
df = pd.read_csv(file_path)

# Check the first few rows of the DataFrame
print(df.head())
```

## 2. Text Preprocessing:

Clean and preprocess the text data in the headlines to remove stop words, punctuation, and perform stemming or lemmatization.

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Tokenize the text
    words = word_tokenize(text)
    # Remove stop words and punctuation, then lemmatize
    words = [lemmatizer.lemmatize(word) for word in words if word.isalpha() and word not in stop_words]
    return ' '.join(words)

# Apply preprocessing to the 'headline' column
df['cleaned_headlines'] = df['headline'].apply(preprocess_text)
```

## 3. Tokenization:

Tokenize the cleaned text into individual words.

```python
vectorizer = CountVectorizer(max_df=0.9, min_df=2, stop_words='english')
```

## 4. Create Document-Term Matrix:

Convert the tokenized text into a document-term matrix (DTM).

```python
dtm = vectorizer.fit_transform(df['cleaned_headlines'])
```

## 5. Apply Topic Modeling:

Use Latent Dirichlet Allocation (LDA) to identify topics within the text data.

```python
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(dtm)
```

## 6. Analyze and Visualize Topics:

Analyze the topics generated and visualize them using word clouds or bar charts.

```python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % (topic_idx))
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

no_top_words = 10
display_topics(lda, vectorizer.get_feature_names(), no_top_words)

# Visualize topics using word cloud
for topic_idx, topic in enumerate(lda.components_):
    print(f"Word Cloud for Topic {topic_idx}")
    wordcloud = WordCloud()
    wordcloud.generate_from_frequencies(dict(zip(vectorizer.get_feature_names(), topic)))
    plt.figure(figsize=(10, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Word Cloud for Topic {topic_idx}")
    plt.show()
```

## 7. Summary of Results
