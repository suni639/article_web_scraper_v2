# Word Frequency Analysis

## Introduction
Using CSV output of the web scraper (article_scraper.ipynb), this simple script analyzes which topics or keywords appear most frequently in the articles. This is performed by preprocessing the text,tokenizing each word or phrase, and removing common words (aka stop words) and punctuation.

The remaining tokenized words can be analyzed, in this case by frequency, and visualized by a bar chart and word cloud.

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

## 1. Load the Data:

Reads the CSV file into a pandas DataFrame.

```python
# Step 1: Load the CSV Data
file_path = 'scraped_articles.csv'
df = pd.read_csv(file_path)

# Check the first few rows of the DataFrame
print(df.head())
```

## 2. Text Preprocessing:

1. Convert to Lowercase: Ensures uniformity in text.
2. Tokenize: Splits text into individual words.
3. Remove Stop Words and Punctuation: Excludes common words like 'the', 'and', etc., using NLTK's stop words list and filtering out non-alphabetic tokens.

```python
# Step 2: Text Preprocessing
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Tokenize the text
    words = word_tokenize(text)
    # Remove stop words and punctuation
    words = [word for word in words if word.isalpha() and word not in stop_words]
    return words

# Apply preprocessing to the 'headline' column
df['cleaned_headlines'] = df['headline'].apply(preprocess_text)
```

## 3. Tokenization and Frequency Analysis:

1. Tokenizes cleaned text into individual words.
2. Counts the frequency of each word using NLTK's FreqDist.

```python
# Step 3: Tokenization and Frequency Analysis
all_words = [word for tokens in df['cleaned_headlines'] for word in tokens]
freq_dist = FreqDist(all_words)

# Display the 20 most common words
print(freq_dist.most_common(20))
```

## 4. Visualization:

1. Creates a bar chart to show the top 20 most common words.
2. Generates a word cloud to visualize word frequencies.

```python
# Step 5: Visualization
# Bar chart
top_words = freq_dist.most_common(25)
words, counts = zip(*top_words)

plt.figure(figsize=(10, 8))
plt.bar(words, counts)
plt.title('Top 20 Most Common Words in Article Headlines')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()


```

```python
# Word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(freq_dist)

plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Article Headlines')
plt.show()
```
